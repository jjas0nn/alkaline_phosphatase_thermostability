finding a linear relationship between initial velocity and [A.P] in tris

```python
import csv
import sys
sys.path.insert(0, 'process_csv/')
from process_csv import *
import matplotlib.pyplot as plt
from scipy.stats import linregress
import requests
with open("dataset/winter_q_summary.csv") as file:
    reader=csv.reader(file)
    data=list(reader)
header=data[0]
i=0
j=1
csvdict={}
while i<len(data[j]):
    csvdict[header[i]]=[]
    while j<len(data):
        csvdict[header[i]].append(data[j][i])
        j+=1
    j=1
    i+=1
for key,val in csvdict.items():
    val=list(map(float, [a for a in val if a]))
    csvdict[key]=val
yval=[]
for i in csvdict[header[2]]:
    yval.append(i*60)
coordict={}
for a,b in zip(csvdict[header[1]],yval):
    #coordict[a]=b
    print(a, ",", b)    
plt.style.use('seaborn-whitegrid')
plt.xlabel("[A.P] in tris mg/mL")
plt.ylabel("initial velocity /min")
plt.title("A.P activity at various A.P concentrations")
plt.rcParams["figure.figsize"] = [20,12]
plt.rcParams.update({'font.size': 22})
plt.plot(csvdict[header[1]],yval,"o")
linregress(csvdict[header[1]],yval)
```

There appears to be a strong linear correlation between [A.P] in tris and initial velocity.
