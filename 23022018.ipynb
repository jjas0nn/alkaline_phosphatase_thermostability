```python
import csv
import sys
sys.path.insert(0, 'process_csv/')
from process_csv import *
import matplotlib.pyplot as plt
import requests
with open("dataset/ap_23022018.csv") as file:
    reader=csv.reader(file)
    data=list(reader)
header=data[0]
i=0
j=1
csvdict={}
while i<len(data[j]):
    csvdict[header[i]]=[]
    while j<len(data):
        csvdict[header[i]].append(data[j][i])
        j+=1
    j=1
    i+=1
for key,val in csvdict.items():
    val=list(map(float, [a for a in val if a]))
    csvdict[key]=val
plt.plot(csvdict[header[1]],label=header[1])
plt.plot(csvdict[header[2]],label=header[2])
plt.plot(csvdict[header[3]],label=header[3])
plt.xlabel("Time sec")
plt.ylabel("PNP absorbance @ 410 nm")
plt.title("A.P activity at various A.P concentrations")
plt.rcParams["figure.figsize"] = [20,12]
plt.legend()
```

## Summary
***
As expected, there appears to be a linear relationship between A.P concentration and initial velocity. Constant freeze/thaw probably destroyed smaller A.P aliquots, explaining why A.P activity in prior experiments was consistently low. Larger A.P aliquots were rarely used and therefore subject to less freeze/thaw.
