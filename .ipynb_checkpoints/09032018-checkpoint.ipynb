```python
import csv
import sys
sys.path.insert(0, 'process_csv/')
from process_csv import *
import matplotlib.pyplot as plt
import requests
with open("dataset/ap_09032018.csv") as file:
    reader=csv.reader(file)
    data=list(reader)
header=data[0]
i=0
j=1
csvdict={}
while i<len(data[j]):
    csvdict[header[i]]=[]
    while j<len(data):
        csvdict[header[i]].append(data[j][i])
        j+=1
    j=1
    i+=1
for key,val in csvdict.items():
    val=list(map(float, [a for a in val if a]))
    csvdict[key]=val
for i in range(1,9):
    plt.plot(csvdict[header[i]],label=header[i])
plt.xlabel("Time sec")
plt.ylabel("PNP absorbance @ 410 nm")
plt.title("A.P activity at various A.P concentrations")
plt.rcParams["figure.figsize"] = [20,12]
plt.legend(prop={'size': 15})
```

```python
import csv
import sys
from scipy.stats import linregress
sys.path.insert(0, 'process_csv/')
from process_csv import *
import matplotlib.pyplot as plt
import requests
xval=[1,1,.5,.5,.25,.25,.125,.125]
yval=[.023,.022,.014,.011,.004,.004,.002,.001]
y=[]
for i in yval:
    y.append(60*i)
plt.plot(xval,y)
plt.xlabel("A.P concentration in tris mg/mL")
plt.ylabel("initial velocity /min")
plt.title("A.P activity due to concentration")
plt.rcParams["figure.figsize"] = [16,12]
linregress(xval,y)
```

## Summary
***
Again, there appears to be a linear relationship between A.P concentration and initial velocity as expected. Smaller aliquots were destroyed due to constant freeze/thaw which explains low activity.
