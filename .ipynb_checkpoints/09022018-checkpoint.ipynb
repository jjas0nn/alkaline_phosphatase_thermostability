# 09022018 kinetic spec assay w/PNPP variable

Since initial velocity was low for both concentrations and did not change like expected, we ran spec assays with different PNPP concentrations but held A.P concentration constant at 5 mg/mL to determine if PNPP was the limiting reagent.

### specs
***
* **measured at normal, 2x, and 10x PNPP concentration** to detect any relationship between PNPP concentration and reaction rate
* **5 mg/mL A.P concentration held constant**
* no osmolytes
* no incubation
* samples measured at time of setup
* used 10mL A.P dilutions from 2015

reagent | final concentration | stock amount
--- | --- | ---
50 mM A.P @ -20 deg C | 5 mg/mL (1/10 dilution) in tris | 200 uL
PNPP | 50 mg/mL | 200 uL
2x PNPP | 100 mg/mL | 200 uL
10x PNPP | 500 mg/mL | 200 uL

```python
import csv
import sys
sys.path.insert(0, 'process_csv/')
from process_csv import *
import matplotlib.pyplot as plt
import requests
with open("dataset/ap_09022018.csv") as file:
    reader=csv.reader(file)
    data=list(reader)
header=data[0]
i=0
j=1
csvdict={}
while i<len(data[j]):
    csvdict[header[i]]=[]
    while j<len(data):
        csvdict[header[i]].append(data[j][i])
        j+=1
    j=1
    i+=1
for key,val in csvdict.items():
    val=list(map(float, [a for a in val if a]))
    csvdict[key]=val
print(csvdict)
plt.plot(csvdict[header[2]],csvdict[header[3]],label=header[3])
plt.plot(csvdict[header[0]],csvdict[header[1]],label=header[1])
plt.plot(csvdict[header[4]],csvdict[header[5]],label=header[5])
plt.plot(csvdict[header[6]],csvdict[header[7]],label=header[7])
plt.xlabel("Time sec")
plt.ylabel("PNP absorbance @ 410 nm")
plt.title("Effect of PNPP concentration on reaction rate")
plt.legend()
```

PNPP concentration | # runs

500mg/mL | 2


### Summary
***
Initial velocity is equally large for all PNPP concentrations. Reaction system is already saturated with PNPP at 50 mg/mL so increasing PNPP concentration has no effect on reaction rate. PNPP is not the limiting reagent so it must be A.P. 5 mg/mL A.P w/ 50mg/mL PNPP initial velocity inconsistent with previous initial velocity (.004 and .124) with same specs.
